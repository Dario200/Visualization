```python
#Change column names
df_MTG_DEL_1 = df_MTG_DEL.add_prefix("MTG_DEL_")
df_CC_DEL_1 = df_CC_DEL.add_prefix("CC_DEL_")
df_Adv_1 = df_Adv.add_prefix("ADV_")
df_REG_1 = df_REG.add_prefix("REG_")

```

```python
# Joins to get output file for Visualization

df_MTG_01 = df_MTG.merge(df_MTG_DEL_1, how='left', left_on='CUST_Id', right_on='MTG_DEL_CUST_Id')
df_MTG_02 = df_MTG_01.merge(df_CC_DEL_1, how='left', left_on='CUST_Id', right_on='CC_DEL_CUST_Id')
df_MTG_03 = df_MTG_02.merge(df_Adv_1, how='left', left_on='Finadvisor', right_on='ADV_FinAdvID')
df_MTG_04 = df_MTG_03.merge(df_REG_1, how='left', left_on='ADV_Province', right_on='REG_Province')
df_MTG_04.head()
```

```python
# Save the DataFrame to an Excel file
df_MTG_04.to_csv("data_for_viz.csv", index=False)
files.download('data_for_viz.csv')
```
